```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
optimal_p = .35
optimal_c = .65
```

```python
def scaling(heat, n):
    a = -30
    h = heat
    k = 1

    return np.max([0, a * (h - n) ** 2 + k])
```

```python
x = np.linspace(0,1,100)
p_curve = [scaling(optimal_p, i) for i in x]
c_cruve = [scaling(optimal_c, i) for i in x]
```

```python
sns.set_theme()
plt.suptitle('Heat Curves', fontweight='bold')
plt.xlabel('Heat Scaling')
plt.ylabel('Percent Efficacy')
plt.ylim((0,1.4))
plt.plot(x, p_curve, c='g', label='Photosynthetic Enzymes')
plt.plot(x, c_cruve, c='b', label='Chemosynthetic Enzymes')
plt.legend()

# reformat legend
```

```python
# 2d feature vectors, (example)

vectors1 = [(np.random.uniform()/10+.2, np.random.uniform()/5+.5) for i in range(50)]
vectors2 = [(np.random.uniform()/5+.5, np.random.uniform()/10+.2) for i in range(50)]

vectormean1 = (np.mean([i[0] for i in vectors1]), np.mean([i[1] for i in vectors1]))
vectormean2 = (np.mean([i[0] for i in vectors2]), np.mean([i[1] for i in vectors2]))

z1 = [np.sqrt((i[0]-vectormean1[0])**2 + (i[1]-vectormean1[1])**2) for i in vectors1]
z2 = [np.sqrt((i[0]-vectormean2[0])**2 + (i[1]-vectormean2[1])**2) for i in vectors2]
```

```python
plt.title('Vectors Example')
plt.xlim((0,1))
plt.ylim((0,1))
plt.scatter([i[0] for i in vectors1], [i[1] for i in vectors1], c=z1, cmap='summer', label='1')
plt.scatter([i[0] for i in vectors2], [i[1] for i in vectors2], c=z2, cmap='autumn', label='2')
plt.legend()
plt.show()
```
